This notebook summarizes how we started the project and chose initial data.

We first downloaded our data set, the munich airbnb `listings.csv.gz`.
Looking at the features we extracted a few features we thought would be able to lead to improved price predictions.
We called them the "relevant features".
They include:



```python
LEGACY_RELEVANT_COLUMNS = [
    "id",
    "review_scores_rating",
    "number_of_reviews",
    "name",
    "description",
    "neighborhood_overview",
    "host_since",
    "host_response_time",
    "host_response_rate",
    "latitude",
    "longitude",
    "bathrooms",
    "bedrooms",
    "beds",
    "price",
    "minimum_nights",
    "maximum_nights",
]
```

After that we engineered a feature, the distance to the Marienplatz, munichs city center, we called the feature `distance_km`.
It uses haversine distance of the latitude and longitude features.

That feature set is saved under `data/processed/legacy.csv`.

After that we created two models testing our features.
A MLP model and a linear model.

Since a lot of columns were still missing, we used only the features which had low missing counts.

First we created a linear model using only the "dinstance_km" feature, to see how well it would perform.

For the MLP (v1) we used ["distance_km",  "review_scores_rating", "number_of_reviews"].

See `legacy_notebooks/Data-Cleaning` for how the data was actually processed for those legacy models.
